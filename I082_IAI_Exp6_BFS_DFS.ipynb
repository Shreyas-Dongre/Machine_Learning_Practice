```python
graph = {
    "A":["B","C"],
    "B":["D","E"],
    "C":["F"],
    "D":[],
    "E":["F"],
    "F":[]
}
```

#Breadth First Search

```python
visited = [] #Keep track of visited nodes
queue = [] #Keep track of nodes in queue
def BFS(graph,visited,node):
  visited.append(node)
  queue.append(node)
  while queue != []:
    s = queue.pop(0) #First In First Out(FIFO) in queue
    print(s,end = " ")
    for neighbour in graph[s]:
      if neighbour not in visited:
        visited.append(neighbour)
        queue.append(neighbour)
```

```python
BFS(graph,visited,"A")
```

#Depth First Search

```python
visited = []
def dfs(visited,graph,node):
  if node not in visited:
    print(node, end = " ")
    visited.append(node)
  for neighbour in graph[node]:
    dfs(visited,graph,neighbour)
```

```python
dfs(visited,graph,"A")
```
