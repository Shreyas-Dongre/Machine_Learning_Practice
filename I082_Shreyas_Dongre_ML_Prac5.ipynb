```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import pandas as pd
import statsmodels.api as sm
import math
from sklearn.datasets import make_blobs
```

```python
#K-Means Algorithm
dataset,classes=make_blobs(n_samples=200,n_features=2,centers=4,cluster_std=0.5,random_state=1)
```

```python
df=pd.DataFrame(dataset,columns=['var1','var2'])
```

```python
df.head()
```

```python
df.shape
```

```python
plt.scatter(df['var1'],df['var2'],color='red')
```

```python
dataset,classes=make_blobs(n_samples=400,n_features=2,centers=4,cluster_std=3,random_state=1)
```

```python
df2=pd.DataFrame(dataset,columns=['var1','var2'])
```

```python
plt.scatter(df2['var1'],df2['var2'],color='red')
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans=KMeans(3)
```

```python
kmeans.fit(df)
```

```python
kmeans.cluster_centers_
```

```python
kmeans.n_iter_
```

```python
kmeans.labels_
```

```python
#color the datapoints acc to clusters they belong
three_clusters=kmeans.fit_predict(df)
df_with_clusters=df.copy()
df_with_clusters['clusters']=three_clusters

plt.scatter(df_with_clusters['var1'],df_with_clusters['var2'],c=df_with_clusters['clusters'],cmap='rainbow')

```

```python
wcss=[]          #within cluster similarity score
for i in range(1,10):
  kmeans=KMeans(i)
  kmeans.fit(df)
  wcss_iter=kmeans.inertia_
  wcss.append(wcss_iter)
```

```python
number_clusters=range(1,10)
plt.plot(number_clusters,wcss)
plt.title('The Elbow')
plt.xlabel('No. of Clusters')
plt.ylabel('WCSS')
```

```python
#implement kmeans on iris dataset and determine optimum no of clusters for it.
df=pd.read_csv("Iris.csv")
df.head()
```

```python
wcss=[]          #within cluster similarity score
for i in range(1,10):
  kmeans=KMeans(i)
  kmeans.fit(df)
  wcss_iter=kmeans.inertia_
  wcss.append(wcss_iter)
```

```python
number_clusters=range(1,10)
plt.plot(number_clusters,wcss)
plt.title('The Elbow')
plt.xlabel('No. of Clusters')
plt.ylabel('WCSS')
```
