```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv('seeds.csv')
```

```python
df.head()
```

```python
df["Type"].value_counts()
```

```python
df.describe()
```

```python
#matplotlib #plot
df.plot(kind="scatter", x="Kernel.Length", y="Kernel.Width")
```

```python
#matplotlib #plot
df.plot(kind="scatter", x="Area", y="Perimeter")
```

```python
#use of seaborn library
sns.jointplot(x="Perimeter",y="Compactness",data=df,size=5)
```

```python
#plot the three types of seeds with three diff colors
sns.FacetGrid(df,hue="Type", size=5).map(plt.scatter,"Perimeter","Compactness").add_legend()
```

```python
#boxplot
sns.boxplot(x='Type',y='Compactness',data=df)
```

```python
ax=sns.boxplot(x="Type",y="Compactness",data=df)
ax=sns.stripplot(x="Type",y="Compactness",data=df,jitter=True,edgecolor="gray")
```

```python
#violin plot
sns.violinplot(x='Type',y='Compactness',data=df,size=5)
```

```python
#kdeplot
sns.FacetGrid(df,hue="Type",size=5).map(sns.kdeplot,"Compactness").add_legend()
```

```python
#pairplot
sns.pairplot(df,hue='Type',size=5,diag_kind='hist')
```

```python
#Andrews curves
from pandas .plotting import andrews_curves
andrews_curves(df,'Type')
```
