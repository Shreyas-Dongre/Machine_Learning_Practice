```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
#TASK-1
x=np.array([5,15,25,35,45,55])
y=np.array([11,16,18,30,22,38])
```

```python
plt.scatter(x,y)
```

```python
#creating regression model
model=LinearRegression()
```

```python
#fit the model on data
#making it 2D cause model wont work on 1D
x=x.reshape(-1,1)
```

```python
x.shape
```

```python
model.fit(x,y)
```

```python
r_sq=model.score(x,y)
print(r_sq)
```

```python
print("Intercept: ",model.intercept_)
```

```python
print("Slope: ",model.coef_)
```

```python
#prediction using the model
x_predict=np.array([20])
x_predict=x_predict.reshape((-1,1))
y_predict=model.predict(x_predict)
print(y_predict)
```

```python
#take x_predict as OG array x and predict the output
x_predict=x
y_predict=model.predict(x_predict)
print(y_predict)
```

```python
#Normalizing the regression line on the data

plt.scatter(x,y)
plt.plot(x_predict,y_predict,"r-")
plt.xlabel("Independent variable X-->")
plt.ylabel("Independent variable Y-->")
plt.title("Simple Linear Regression")
```
