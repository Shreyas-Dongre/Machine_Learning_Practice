```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import pandas as pd
import statsmodels.api as sm
import math
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_blobs
```

```python
dataset,classes=make_blobs(n_samples=200,n_features=2,centers=1,cluster_std=0.5,random_state=1)
```

```python
plt.scatter(dataset[:,0],dataset[:,1])
```

```python
classes
```

```python
dbscan=DBSCAN(eps=0.3,min_samples=20)
print (dbscan)
```

```python
pred=dbscan.fit_predict(dataset)
pred
```

```python
#visualising outliers and clusters

#sample no of outliers


outlier_index=np.where(pred==-1)

outlier_values = dataset[outlier_index]
```

```python
plt.scatter(dataset[:,0],dataset[:,1])
plt.scatter(outlier_values[:,0],outlier_values[:,1],color='r')
```

```python
from sklearn.datasets import make_circles
dataet,classes=make_circles(n_samples=500,factor=0.3,random_state=1)
```

```python
dataset,classes = make_circles(n_samples = 500, factor = 0.3, random_state = 1)
plt.scatter(dataset[:,0],dataset[:,1],c=classes)
```

```python
dataset,classes = make_circles(n_samples = 500, factor = 0.3, random_state = 1,noise=0.1)
plt.scatter(dataset[:,0],dataset[:,1],c=classes)
```

```python
dbscan=DBSCAN(eps=0.2,min_samples=20)
pred=dbscan.fit_predict(dataset)
outlier_index=np.where(pred==-1)
outlier_values = dataset[outlier_index]
plt.scatter(dataset[:,0],dataset[:,1],c=classes)
plt.scatter(outlier_values[:,0],outlier_values[:,1],color='r')
```

```python
from sklearn.datasets import make_moons
dataset,classes = make_moons(n_samples = 500,noise=0.15, random_state = 1)
plt.scatter(dataset[:,0],dataset[:,1],c=classes)
```

```python
dbscan=DBSCAN(eps=0.169,min_samples=20)
pred=dbscan.fit_predict(dataset)
outlier_index=np.where(pred==-1)
outlier_values = dataset[outlier_index]
plt.scatter(dataset[:,0],dataset[:,1],c=classes)
plt.scatter(outlier_values[:,0],outlier_values[:,1],color='r')
```

```python
#Implement DBSCAN on Dabetes dataset.
df=pd.read_csv("diabetes.csv")
df.head()
```

```python
plt.scatter(df['BMI'],df['Glucose'])
```

```python
df=df.to_numpy()
```

```python
dbscan=DBSCAN(eps=15,min_samples=180)
pred=dbscan.fit_predict(df)
outlier_index=np.where(pred==-1)
outlier_values = df[outlier_index]
plt.scatter(df[:,0],df[:,1])
plt.scatter(outlier_values[:,0],outlier_values[:,1],color='r')
```
